```python
# Список регионов
region_list = [
    'Алтайский край', 'Амурская область', 'Архангельская область', 'Астраханская область', 'Белгородская область',
    'Брянская область', 'Владимирская область', 'Волгоградская область', 'Вологодская область', 'Воронежская область',
    'Москва', 'Санкт-Петербург', 'Севастополь', 'Еврейская автономная область', 'Забайкальский край',
    'Ивановская область', 'Иркутская область', 'Кабардино-Балкарская Республика', 'Калининградская область', 'Калужская область',
    'Камчатский край', 'Карачаево-Черкесская Республика','Кемеровская область','Кировская область',
    'Костромская область', 'Краснодарский край', 'Красноярский край', 'Курганская область', 'Курская область',
    'Ленинградская область', 'Липецкая область', 'Магаданская область', 'Московская область', 'Мурманская область',
    'Ненецкий АО', 'Нижегородская область', 'Новгородская область', 'Новосибирская область', 'Омская область',
    'Оренбургская область', 'Орловская область', 'Пензенская область', 'Пермский край', 'Приморский край',
    'Псковская область', 'Республика Адыгея', 'Республика Алтай', 'Республика Башкортостан', 'Республика Бурятия',
    'Республика Дагестан', 'Республика Ингушетия', 'Республика Калмыкия', 'Республика Карелия', 'Республика Коми',
    'Республика Крым', 'Республика Марий Эл', 'Республика Мордовия', 'Республика Саха (Якутия)',
    'Республика Северная Осетия-Алания', 'Республика Татарстан', 'Республика Тыва', 'Республика Хакасия',
    'Ростовская область', 'Рязанская область', 'Самарская область', 'Саратовская область', 'Сахалинская область',
    'Свердловская область', 'Смоленская область', 'Ставропольский край', 'Тамбовская область', 'Тверская область',
    'Томская область', 'Тульская область', 'Тюменская область', 'Удмуртская Республика', 'Ульяновская область',
    'Хабаровский край', 'Ханты-Мансийский АО', 'Челябинская область', 'Чеченская Республика', 'Чувашская Республика',
    'Чукотский АО', 'Ямало-Ненецкий АО','Ярославская область'
    ]

# Количество регионов
len_region_list = len(region_list)

# Функция обработки названий регионов
def rename_regions(region_name):
    if 'автономный округ' in region_name:
        region_name = region_name.replace('автономный округ', 'АО')
    if 'авт. округ' in region_name \
        and "без авт. округа" not in region_name\
        and "без авт. округов" not in region_name:
        region_name = region_name.replace('авт. округ', 'АО')
    if 'авт. область' in region_name:    
        region_name = region_name.replace('авт. область', 'автономная область')
    if "округ" in region_name \
        and "без автономного округа" not in region_name \
        and "без автономных округов" not in region_name \
        and "без авт. округа" not in region_name\
        and "без авт. округов" not in region_name\
        and "кроме" not in region_name:
        return None
    if "федеральный округ" in region_name:
        return None
    if "район" in region_name:
        return None
    if "сельсовет" in region_name:
        return None
    if "поссовет" in region_name:
        return None
    if "рп Благовещенка" in region_name:
        return None
    if "рп Степное Озеро" in region_name:
        return None
    if "Российская Федерация" in region_name:
        return None  
    if "в том числе:" in region_name:
        return None  
    if "Чеченская и Ингушская Республики" in region_name:
        return None   
    if "Байконур" in region_name:
        return None
    if "Раздел" in region_name:
        return None      
    if "Москва" in region_name:
        return "Москва"
    if "Санкт-Петербург" in region_name:
        return "Санкт-Петербург"
    if "Санкт - Петербург" in region_name:
        return "Санкт-Петербург"
    if "Долгано-Ненецкий" in region_name:
        return None
    if "Ямало-Ненецкий" in region_name:
        return "Ямало-Ненецкий АО"
    if "Ненецкий" in region_name and "Ямало-Hенецкий" not in region_name:
        return "Ненецкий АО"
    if "Ханты-Мансийский" in region_name:
        return "Ханты-Мансийский АО"
    if "Чувашская" in region_name:
        return "Чувашская Республика"
    if 'Татарстан' in region_name:
        return "Республика Татарстан"
    if "Крым" in region_name:
        return "Республика Крым"
    if 'Севастополь' in region_name:
        return "Севастополь"  
    if 'Кемеровская' in region_name:
        return 'Кемеровская область'
    if 'Адыгея' in region_name:
        return 'Республика Адыгея'
    if "Осетия – Алания" in region_name:
        return "Республика Северная Осетия-Алания"
    if "Осетия-Алания" in region_name:
        return "Республика Северная Осетия-Алания"
    if "Осетия - Алания" in region_name:
        return "Республика Северная Осетия-Алания"
    if "Эвенкийский" in region_name:
        return None 
    if "Республика Саха(Якутия)" in region_name:
        return "Республика Саха (Якутия)"
    if "Архангельская область без автономного округа" in region_name \
        or "Архангельская область без авт. округа" in region_name \
        or "Архангельская область (кроме Ненецкого автономного округа)" in region_name \
        or "Архангельская область (без АО)" in region_name:
        return "Архангельская область"
    if "Тюменская область без автономных округов" in region_name \
        or "Тюменская область без автономного округа" in region_name \
        or "Тюменская область без авт. округов" in region_name \
        or "Тюменская область (без АО)" in region_name \
        or "Тюменская область (кроме Ханты-Мансийского автономного округа-Югры и Ямало-Ненецкого автономного округа)" in region_name:
        return "Тюменская область"
    if "Главное медицинское управление Управления делами Президента Российской Федерации" in region_name:
        return None
    if "Муниципальные образования Алтайского края" in region_name:
        return None    
    if "Населенные пункты, входящие в состав муниципальных образований Алтайского края" in region_name:
        return None
    if "Московская обл. в старых границах" in region_name:
        return None
         
    return region_name

# Функция переименования наименования столбцов с годом типа float в object
def float_to_object(df):
    df.rename(columns={
        2015.0: '2015', 
        2016.0: '2016', 
        2017.0: '2017', 
        2018.0: '2018',
        2019.0: '2019',
        2020.0: '2020',
        2021.0: '2021',
    }, inplace=True)
    return df
 
# Функция удаления строк с Архангельской и Тюменской областями
def exclude_Arhangelsk_Tumen_region(df):
    df = df.loc[~df['region'].isin([
        'Архангельская область',
        'Тюменская область'
    ])]
    return df 

# Функция обработки столбца с наименованием регионов
def region_name_columns_processing(df, exclude_A_T=True):
    '''Удаляет лишний перенос строки,
    удаляет лишние пробелы в начале и конце названия регионов,
    удаляет строки с наименованием федеральных округов и обобщенные данные,
    вносит изменения в названия регионов,
    изменяет тип типизацию колонок float -> object'''

    # Удаляем лишний перенос строк
    df['region'].replace(r'\s+|\\n', ' ', regex=True, inplace=True)
    # удалим лишние пробелы в названиях регионов
    df['region'] = df['region'].str.strip()
    # Удаляем строки с наименованием Архангельской и Тюменской областей
    if exclude_A_T:
        df = exclude_Arhangelsk_Tumen_region(df)
    # Переименовываем названия регионов
    df.region = df.region.apply(rename_regions)
    # Производим переименование колонок
    df = float_to_object(df)
    # Удаляем строки с NaN в наименованиях регионов
    df = df.dropna(subset = ['region']) 

    return df

# Функция для проверки соответствия изменений в названиях регионов шаблонному списку
def change_control(*args):
    for arg in args:
        if len(set(region_list).difference(set(arg.region))) != 0:
            print('Множества регионов не совпадают. Перечень несовпадений:')   
            print(set(region_list).difference(set(arg.region)))
        if arg.shape[0] != len_region_list:
            print('Размерности не совпадают,', arg.shape[0], 'вместо', len_region_list)
        if len(set(region_list).difference(set(arg.region))) == 0 and arg.shape[0] == len_region_list:
            print('Done!')
            
# Функция обработки датасетов по детской смертности
def child_mortality(df):
    df.rename(columns={
        'Unnamed: 0': 'region', 
        'Unnamed: 1': 'to_delete', 
        'Unnamed: 2': 'type_of_population', 
    }, inplace=True)
    df.drop('to_delete', axis=1, inplace=True)
    return df


# Функция обработки датасета по рождаемости       
def newborn_monthly_processing(df):
    df.rename(columns={
        'Region': 'region', 
        'Unnamed: 198': 'to_delete', 
    }, inplace=True)
    df.drop('to_delete', axis=1, inplace=True)
    df = df.iloc[:-2]
    for col_name in df.columns[1:]:
        df.rename(columns={col_name: col_name[-7:-3]}, inplace=True)
    return df

# Функция преобразования датасета по рождаемости  
def newborn_monthly_conversion(df):
    df = df.fillna(0)
    for col_name in df.columns[1:]:
        df[col_name] = df[col_name].apply(
            lambda x: x.astype(str)\
                        .str.replace(',',  '.')\
                        .astype(float)\
                        .round()\
                        .astype(int)
            )
    df = df.set_index('region').transpose().reset_index()
    df = df.groupby(by=df['index'])[df.columns[1:]].sum()
    df = df.transpose().reset_index()
    return df

# Функция обработки датасета по численности населения
def population_processing(df):
    df.rename(columns={
        'Unnamed: 0': 'region', 
        'Unnamed: 1': 'to_delete', 
    }, inplace=True)
    df.drop('to_delete', axis=1, inplace=True)
    df = df.iloc[1:]
    for col_name in df.columns[1:]:
        df.rename(columns={col_name: col_name[-7:-3]}, inplace=True)
    df = df.set_index('region').shift(periods=-1).reset_index()
    return df

# Функция обработки датасета по обороту розничной торговли на душу населения
def retail_turnover_per_capita_processing(df):
    df.rename(columns={
        'Unnamed: 0': 'region', 
        'Unnamed: 1': 'to_delete', 
    }, inplace=True)
    df.drop('to_delete', axis=1, inplace=True)
    df['region'] = df['region'].str.strip()
    df = df.loc[~df['region'].isin([
    'Архангельская область (без АО)',
    'Тюменская область (без АО)'
    ])]
    df = df.iloc[1:]
    return df

# Функция обработки датасета занятости населения
def workers_processing(df):
    df.rename(columns={
        'Unnamed: 0': 'to_delete', 
        'Unnamed: 1': 'region', 
    }, inplace=True)
    df.drop('to_delete', axis=1, inplace=True)
    df.loc[26, ['2012', '2013', '2014']] = df.loc[25, ['2012', '2013', '2014']]
    df.loc[74, ['2012', '2013', '2014']] = df.loc[73, ['2012', '2013', '2014']]
    df['region'] = df['region'].str.strip()
    df = df.loc[~df['region'].isin([
    'Архангельская область (без АО)',
    'Тюменская область (без АО)'
    ])]
    return df

# Функция обработки датасетов распределения населения по социально-демографическим группам
def poverty_socdem_processing(*args):
    poverty_socdem_child = pd.DataFrame(data=region_list, columns=['region'])
    poverty_socdem_adult = pd.DataFrame(data=region_list, columns=['region'])
    poverty_socdem_elderly = pd.DataFrame(data=region_list, columns=['region'])
    for arg in args:
        poverty_socdem_child = poverty_socdem_child.merge(
            arg.iloc[:,:2],
            on='region',
            how='inner'
        )
        poverty_socdem_adult = poverty_socdem_adult.merge(
            arg.iloc[:,[0, 3]],
            on='region',
            how='inner'
        )
        poverty_socdem_elderly = poverty_socdem_elderly.merge(
            arg.iloc[:,[0, 2]],
            on='region',
            how='inner'
        )
    return poverty_socdem_child, poverty_socdem_adult, poverty_socdem_elderly

# Функция переименования столбцов:
def rename_col(df):
    for name in df.columns[1:]:
        df.rename(columns={name: name[-4:]}, inplace=True)
    return df

# Функция обработки датасетов отгруженных товаров и услуг №1 - суммирование по разделам:
def regional_production_processing(df_1, df_2):
    total_1 = df_1.drop('production_field', axis=1)
    total_1 = total_1.groupby(by='region').sum().reset_index()
    total_2 = df_2.drop('production_field', axis=1)
    total_2 = total_2.groupby(by='region').sum().reset_index()  
    result = total_1.merge(
       total_2,
       on='region',
       how='inner' 
    )
    return result

# Функция обработки датасета заболеваемости населения:
def morbidity_age_disease_processing(df):
    df = df[df['age'] == 'Всего']
    df.drop('age', axis=1, inplace=True)
    df = df.groupby(by='region').sum().reset_index()
    df.loc[3, ['2014', '2015', '2016']] = df.loc[2, ['2014', '2015', '2016']]
    df.loc[76, ['2014', '2015', '2016']] = df.loc[75, ['2014', '2015', '2016']]
    df = df.loc[~df['region'].isin([
        'Архангельская обл. без данных по Ненецкому авт. окр.',
        'Тюменская обл.без данных по Ханты-Мансийскому и Ямало-Ненецкому авт. окр.'
    ])]
    return df

# Функции обработки датасета о потреблении на душу населения:
def per_capita_household_consumption_processing(df):
    df.rename(columns={
        'Unnamed: 0': 'to_delete',
        'Unnamed: 1': 'region',
    }, inplace=True)
    df.drop('to_delete', axis=1, inplace=True)
    df.loc[26, ['2012', '2013', '2014']] = df.loc[25, ['2011', '2012', '2013', '2014']]
    df.loc[75, ['2012', '2013', '2014']] = df.loc[74, ['2011', '2012', '2013', '2014']]
    df['region'] = df['region'].str.strip()
    df = df.loc[~df['region'].isin([
    'Архангельская область (без АО)',
    'Тюменская область (без АО)'
    ])]
    return df

# Функции обработки датасетов о структуре расходов потребления:
def structure_of_household_expenditure_processing(df):
    df = df.transpose().reset_index()
    df.rename(columns={
        'index': 'region',
        0: 'Всего',
        1: 'Оплата услуг',
        2: 'Покупка алкогольных напитков',
        3: 'Покупка непродовольственных товаров',
        4: 'Покупка продуктов питания',
        5: 'Стоимость услуг, оказанных работодателем бесплатно или по льготным ценам',
    }, inplace=True)
    df = df[1:]
    df.drop('Всего', axis=1, inplace=True)
    return df

# Функции обработки датасета структуры численност инаселения:
def population_structure_by_age_group_processing(df):
    df.rename(columns={
        'Unnamed: 0': 'age_category',
        'Unnamed: 1': 'region',
    }, inplace=True)
    df.drop(['Unnamed: 2', 'Unnamed: 3'], axis=1, inplace=True)
    child_df = df[df['age_category'] == 'Моложе трудоспособного']
    child_df = child_df.reset_index()
    child_df.drop(['index', 'age_category'], axis=1, inplace=True)
    adult_df = df[df['age_category'] == 'Трудоспособное']
    adult_df = adult_df.reset_index()
    adult_df.drop(['index', 'age_category'], axis=1, inplace=True)
    elderly_df = df[df['age_category'] == 'Старше трудоспособного']
    elderly_df = elderly_df.loc[~elderly_df['region'].isin([
        '            Архангельская область (без АО)',
        '            Тюменская область (без АО)'
        ])]
    elderly_df = elderly_df.reset_index()
    elderly_df.drop(['index', 'age_category'], axis=1, inplace=True)
    return child_df, adult_df, elderly_df

# Функции обработки датасетов о преступлениях:
def criminal_processing_2020(df):
    df.rename(columns={
        'Unnamed: 1': 'region',
        'в состоянии опьянения': 'в алкогольном опьянении',
        'Unnamed: 10': 'в наркотическом опьянении',
        'Unnamed: 11': 'в токсическом опьянении',
        'несовершеннолетними или при их соучастии\n': 'несовершеннолетними',
        'ранее совершавшими преступления\n': 'ранее совершавшими преступления'        
    }, inplace=True)
    df['совершенные группой'] = df['группой лиц\n'] + df['группой лиц по предварительному сговору\n'] + \
        df['организованной группой\n'] + df['преступным сообществом (преступной организацией)\n']
    df['в опьянении'] = df['в алкогольном опьянении'] + df['в наркотическом опьянении'] + \
        df['в токсическом опьянении']
    df = df.iloc[2:-9, [1, 2, 3, -2, -1]]
    return df

def criminal_processing_2021(df):
    df.rename(columns={
        'Unnamed: 1': 'region',
        'в состоянии опьянения': 'в алкогольном опьянении',
        'Unnamed: 10': 'в наркотическом опьянении',
        'Unnamed: 11': 'в токсическом опьянении',
        'несовершеннолетними или при их соучастии\n': 'несовершеннолетними',
        'ранее совершавшими преступления\n': 'ранее совершавшими преступления'        
    }, inplace=True)
    df['совершенные группой'] = df['группой лиц'] + df['группой лиц по предварительному сговору'] + \
        df['организованной группой'] + df['преступным сообществом (преступной организацией)']
    df['в опьянении'] = df['в алкогольном опьянении'] + df['в наркотическом опьянении'] + \
        df['в токсическом опьянении']
    df = df.iloc[2:-9, [1, 2, 3, -2, -1]]
    return df

# Функция отображения пропусков в обработанных данных
def view_null_data(df):
    null_data = df.isnull().sum()
    if null_data.sum() > 0:
        cols_with_null = null_data[null_data > 0].sort_values(ascending=False)
        regions_with_null = df[df.isnull().any(axis = 1)]
        print('Колонки с отсутсвующими значениями:\n', cols_with_null)
        print('Регионы с отсутсвующими значениями:\n', regions_with_null)
        cols_with_null.plot(
            kind='bar',
            figsize=(10, 4),
            title='Количество пропусков в данных'
            )
    else:
        print('Пропусков нет')

# Функция обработки пропусков
def nan_processing(df, is_mean=False, is_median=False):
    if is_mean:
        df = df.apply(lambda row: row.fillna(row[1:].mean()), axis=1)
    if is_median:
        df = df.apply(lambda row: row.fillna(row[1:].median()), axis=1)
    return df

# Функция обработки нулевых значений
def zero_processing(df, col, is_mean=False, is_median=False):
    if is_mean:
        for idx in df[df[col] == 0.0].index:
            df.loc[idx, col] = round(df.iloc[idx, 1:].mean(), 1)
    if is_median:
        for idx in df[df[col] == 0.0].index:
            df.loc[idx, col] = round(df.iloc[idx, 1:].median(), 1)
    return df

# Функция для добавления колонки со средним/медианой по региону
def data_update(df, col_name, year, is_mean=False, is_median=False):
    ''' df - обрабатываемый датафрейм
        col_name - наименование голонки, которая создается в результате работы функции
        year - наименование колонки с которой производится аггрегация'''
    if is_mean:
        for idx in df.index:
            df.loc[idx, col_name] = round(df.loc[idx, year:].mean(), 1)
    if is_median:
        for idx in df.index:
            df.loc[idx, col_name] = round(df.loc[idx, year:].mean(), 1)
    return df

'''Пришлось перенести вниз из-за необходимости нейминга'''
# Функция создания сводного датасета по регионам по одной из колонок в каждом
# def make_total_table(*args, col='2019'):
#     df = pd.DataFrame(data=region_list, columns=['region'])
#     for arg in args:
#         df = df.merge(
#             arg[['region', col]],
#             on='region',
#             how='inner'
#         )
#         df.rename(columns={col: f'{name_dict[arg]}_{col}'}, inplace=True)
#     return df
```

Создадим функцию, возвращающую имя датафрейма для переименования столбцов в функции make_total_table

```python
def dataframe_name(df):
    if df is per_capita_cash_income:
        return 'per_capita_cash_income'
    elif df is real_incomes: 
        return 'real_incomes'
    elif df is formal_wage_paid:
        return 'formal_wage_paid'
    elif df is real_pay: 
        return 'real_pay'
    elif df is child_mortality_urban:
        return 'child_mortality_urban'
    elif df is child_mortality_rural:
        return 'child_mortality_rural'
    elif df is  child_mortality_total:
        return 'child_mortality_total'    
    elif df is disabled_total_by_age_pivot:
        return 'disabled_total_by_age_pivot'
    elif df is alco:
        return 'alco'
    elif df is drugs:
        return 'drugs'
    elif df is alco_drugs_2019:
        return 'alco_drugs_2019'
    elif df is gross_regional_product:
        return 'gross_regional_product'
    elif df is housing_cond:
        return 'housing_cond'
    elif df is housing_intent:
        return 'housing_intent'
    elif df is cramped_living:
        return 'cramped_living'
    elif df is area_per_person:
        return 'area_per_person'
    elif df is improved_housing:
        return 'improved_housing'
    elif df is newborn_yearly:
        return 'newborn_yearly'
    elif df is population:
        return 'population'
    elif df is poverty_percent_by_regions:
        return 'poverty_percent_by_regions'
    elif df is poverty_socdem_child:
        return 'poverty_socdem_child'
    elif df is poverty_socdem_adult:
        return 'poverty_socdem_adult'
    elif df is poverty_socdem_elderly:
        return 'poverty_socdem_elderly'
    elif df is total_goods_and_services:
        return 'total_goods_and_services'
    elif df is retail_turnover_per_capita:
        return 'retail_turnover_per_capita'
    elif df is welfare_expense_share:
        return 'welfare_expense_share'
    elif df is workers:
        return 'workers'
    elif df is morbidity_age_disease:
        return 'morbidity_age_disease'
    elif df is per_capita_household_consumption:
        return 'per_capita_household_consumption'    
    elif df is recidivists:
        return 'recidivists'
```

```python
# Функция создания сводного датасета по регионам по одной из колонок в каждом
def make_total_table(*args, col='2019'):
    df = pd.DataFrame(data=region_list, columns=['region'])
    for arg in args:
        name = dataframe_name(arg)
        df = df.merge(
            arg[['region', col]],
            on='region',
            how='inner'
        )
        df.rename(columns={col: f'{name}_{col}'}, inplace=True)
    return df
```

Сгруппируем данные для изучения регионов, нуждающихся в помощи в первую очередь.
